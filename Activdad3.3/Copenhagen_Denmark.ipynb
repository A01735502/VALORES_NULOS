```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Copenhagen_Denmark.csv')
df.head()
```

```python
visualizaciÃ³n de columnas y filas completas
pd.options.display.max_rows=None
pd.options.display.max_columns=None
```

```python
##creaciÃ³n de copia del dataframe 
data1=df.copy()
```

```python
##visualizaciÃ³n de datos nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
##agrupacion neighbourhood y conteo de registros
data2=data1.groupby(['neighbourhood'])['neighbourhood'].count() 
data2
```

```python
##sustituciÃ³n de valores nulos en la columna neighbourhood por "Denmark, Copenhagen"
data1["neighbourhood"] =data1["neighbourhood"].fillna("Denmark, Copenhagen") 
```

```python
##comprobacion de cambios de valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
##la lÃ³gica de que cada propiedad tenga al menos un baÃ±o, por lo que se reemplazan los valores nulos por 1
data1["bathrooms"] =data1["bathrooms"].fillna(1) 
```

```python
##los valores nulos en la columna "bedrooms" se reemplazan con el valor redondeado hacia arriba de la mitad del valor en la columna "accommodates"(capacidad)
data1.loc[data1["bedrooms"].isna(), "bedrooms"] = np.ceil(data1["accommodates"] / 2)
```

```python
#comprobacion de valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
##meiante la lÃ³gica de que cada propiedad tenga al menos una cama, se reemplazan los valores nulos en la columna "beds" con el valor redondeado hacia arriba de la mitad del valor en la columna "accommodates"(capacidad)
data1.loc[data1["beds"].isna(), "beds"] = np.ceil(data1["accommodates"] / 2)
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
###los valores nulos en las columnas "license", "description" y "neighborhood_overview" se reemplazan por "No tiene licencia", "No tiene descripciÃ³n" y "No tiene descripciÃ³n del vecindario", respectivamente. Porque son campos de texto y es preferible tener una indicaciÃ³n clara de que no hay informaciÃ³n disponible en lugar de dejar el campo vacÃ­o.
data1["license"] =data1["license"].fillna("No tiene licencia")
data1["description"] =data1["description"].fillna("No tiene descripciÃ³n") 
data1["neighborhood_overview"] =data1["neighborhood_overview"].fillna("No tiene descripciÃ³n del vecindario") 
```

```python
##agrupacion room_type y conteo de registros
data2=data1.groupby(['room_type'])['room_type'].count() 
data2 
```

```python
##conversiÃ³n de la columna price a tipo float, eliminando el sÃ­mbolo de dÃ³lar y las comas
data1["price"] = data1["price"].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
##cÃ¡lculo del promedio de la columna price agrupado por room_type
prom_por_tipo = data1.groupby("room_type")["price"].mean().reset_index()
print("Promedio por room_type:\n",prom_por_tipo)
```

```python
###cÃ¡lculo del promedio de la columna price agrupado por room_type y accommodates
prom_por_tipo_bath = (data1.groupby(["room_type", "accommodates"], dropna=False)["price"].mean().reset_index())
print(prom_por_tipo_bath.head(20))
```

```python
###los valores nulos en la columna price se reemplazan por el promedio de la columna price agrupado por room_type
data1["price"] = data1["price"].fillna(
    data1.groupby("room_type")["price"].transform("mean"))
```

```python
# Si no hay dato de superhost/ no es superhost
data1["host_is_superhost"] = data1["host_is_superhost"].fillna("f")

# Reglas si es superhost
data1.loc[data1["host_is_superhost"] == "t", "host_response_time"] = "within a few hours"
data1.loc[data1["host_is_superhost"] == "t", "host_acceptance_rate"] = \
    data1.loc[data1["host_is_superhost"] == "t", "host_acceptance_rate"].fillna(100)

# Reglas si NO es superhost
data1.loc[data1["host_is_superhost"] == "f", "host_response_time"] = "within a few days or more"
data1.loc[data1["host_is_superhost"] == "f", "host_acceptance_rate"] = \
    data1.loc[data1["host_is_superhost"] == "f", "host_acceptance_rate"].fillna(0)
```

```python
# Quitar el "%" si viene en string y convertir a numÃ©rico
data1["host_response_rate"] = data1["host_response_rate"].astype(str).str.replace("%", "")
data1["host_response_rate"] = pd.to_numeric(data1["host_response_rate"], errors="coerce")

# Calcular media ignorando NaN
mean_response_rate = data1["host_response_rate"].mean()

# Rellenar los NaN con la media
data1["host_response_rate"].fillna(mean_response_rate, inplace=True)
    
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
##relleno de valores nulos restantes en columnas relacionadas con el anfitriÃ³n
data1["host_listings_count"] =data1["host_listings_count"].fillna(99)
data1["host_verifications"] =data1["host_verifications"].fillna("no tiene verificaciÃ³n")
data1["host_has_profile_pic"] =data1["host_has_profile_pic"].fillna("no tiene foto de perfil")
data1["host_identity_verified"] =data1["host_identity_verified"].fillna("no estÃ¡ verificado")
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
##relleno de valores nulos en columnas relacionadas con las reseÃ±as,, calculo de el revenue estimado mediante el precio y la ocupaciÃ³n estimada en los Ãºltimos 365 dÃ­as 
data1["first_review"] =data1["first_review"].fillna("01/01/1900")
data1["last_review"] =data1["last_review"].fillna("01/01/2100")
data1["estimated_revenue_l365d"]=(data1["price"]*data1["estimated_occupancy_l365d"])



```

```python
##comprobaciÃ³n de datos 
data1[["price", "estimated_occupancy_l365d", "estimated_revenue_l365d"]].head()
```

```python
##relleno de valores nulos con el valor 0 debido a escala likert de 1 al 5 en columnas relacionadas con las reseÃ±as 
data1["review_scores_rating"] =data1["review_scores_rating"].fillna(0)
data1["review_scores_cleanliness"] =data1["review_scores_cleanliness"].fillna(0)
data1["review_scores_checkin"] =data1["review_scores_checkin"].fillna(0)
data1["review_scores_communication"] =data1["review_scores_communication"].fillna(0)
data1["review_scores_location"] =data1["review_scores_location"].fillna(0)
data1["review_scores_value"] =data1["review_scores_value"].fillna(0)
data1["review_scores_accuracy"] =data1["review_scores_accuracy"].fillna(0)
```
